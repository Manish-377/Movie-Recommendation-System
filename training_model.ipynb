```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = pd.read_csv("dummy.txt",index_col=0)
ratings
```

```python
ratings = ratings.fillna(0)
ratings
```

```python
def standardize(row):#approximation
    new_row = (row - row.mean())/(row.max()-row.min())
    return new_row
ratings_std = ratings.apply(standardize)
print(ratings_std)
```

```python
item_similarity = cosine_similarity(ratings_std.T) #item-item similarity, therefore transposing it, as we have users on the rows
print(item_similarity)
```

```python
corrMatrix = pd.DataFrame(item_similarity,index=ratings.columns,columns=ratings.columns)
corrMatrix
```

```python
corrMatrix = corrMatrix.corr(method='pearson')
corrMatrix
```

```python
def get(movie_name,rating):
    similar_score = corrMatrix[movie_name]*(rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)
    return similar_score
```

```python
print (get("movie1",5))
```

```python
user1 = [("animated1",5),("animated2",1),("movie2",1)]
similar_movies = pd.DataFrame()
for movie,rating in user1:
    similar_movies = similar_movies.append(get(movie,rating),ignore_index=True)
```
