```python
import pandas as pd
```

```python
ratings_df = pd.read_csv('ratings.csv')
ratings_df.head()
```

```python
movies_df = pd.read_csv('movies.csv')
movies_df.head()
```

```python
ratings_df = pd.merge(movies_df,ratings_df,on='movieId')
ratings_df.head(160)
```

```python
ratings_df.drop(['genres','timestamp'],axis=1)
print(ratings_df.shape)
```

```python
ratings_df.head(25)
```

```python
userRatings = ratings_df.pivot_table(index=['userId'],columns=['title'],values='rating')
userRatings.head(30)
```

```python
userRatings = userRatings.dropna(thresh=7, axis=1).fillna(0) #axis=1 coz column is considered 
userRatings.head()
```

```python
userRatings = ratings_df.pivot_table(index=['userId'],columns=['title'],values='rating')
userRatings.head()
print("DATA CONSIDERED BEFORE APPLYING THRESH: ",userRatings.shape)
userRatings = userRatings.dropna(thresh=20, axis=1).fillna(0,axis=1)  #use of dropna
print("DATA CONSIDERED AFTER APPLYING THRESH: ",userRatings.shape)
```

```python
#PEARSON CORREALTION ALGORITHM USED BELOW TO FIND OUT CORRELATION BETWEEN ONE AND ANOTHER MOVIE.
#VALUE OF CORRELATION IS 1 FOR SAME INPUTS ie CORRELATION BETWEEN MOVIE a AND MOVIE a will be unity is 1.
correlation_Matrix = userRatings.corr(method='pearson')    
correlation_Matrix.head(150)
```

```python
def similar_movies_type(movie_name,rating):
    similar_ratings = correlation_Matrix[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    print(type(similar_ratings))
    return similar_ratings
```

```python
romantic_lover = [("(500) Days of Summer (2009)",5),("Alice in Wonderland (2010)",3),("Aliens (1986)",1),("2001: A Space Odyssey (1968)",2)]
similar_movies = pd.DataFrame()
for movie,rating in romantic_lover:
    similar_movies = similar_movies.append(similar_movies_type(movie,rating),ignore_index = True)

similar_movies.head()
```

```python
similar_movies.sum().sort_values(ascending=False).head(30)
```

```python
action_lover = [("Amazing Spider-Man, The (2012)",5),("Mission: Impossible III (2006)",4),("Toy Story 3 (2010)",2)]
similar_movies = pd.DataFrame()
for movie,rating in action_lover:
    similar_movies = similar_movies.append(similar_movies_type(movie,rating),ignore_index = True)

similar_movies.head(10)
similar_movies.sum().sort_values(ascending=False).head(30)
```
